```python
class Persona:
    def __init__(self, nombre, edad, dni):
        self.nombre = nombre
        self.edad = edad
        self.dni = dni

    def leerDatos(self):
        self.nombre = input("Ingrese el nombre: ")
        self.edad = input("Ingrese la edad: ")
        self.dni = input("Ingrese el DNI: ")

    def mostrarDatos(self):
        print("Nombre:", self.nombre)
        print("Edad:  ", self.edad)
        print("DNI:   ", self.dni)

# Fin de la clase Persona

# Creando el primer objeto: persona1
print("Persona Numero 1")
persona1 = Persona("", "", "")
persona1.leerDatos()
persona1.mostrarDatos()

print("--------------------------")

# Creando el segundo objeto: persona2
print("Persona Numero 2")
persona2 = Persona("", "", "")
persona2.leerDatos()
persona2.mostrarDatos()

print("--------------------------")
print("Software desarrollado por Meredith Aguirre NL = 2")
```
